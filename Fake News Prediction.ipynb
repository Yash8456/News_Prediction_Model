## TEXT PROCESSING

```python
def wordopt(text):  #text preprocessing function, commonly used in NLP (Natural Language Processing) before applying ML models.
    text=text.lower()  #Converts all characters to lowercase
    
    text=re.sub('\[.*?\]' , ' ',text) #Removes anything inside square brackets [ ... ].
                                     #Example: "this is [note] text" → "this is text"
    
    text=re.sub("\\W"," ", text) #\W matches any non-word character (anything except letters, digits, and underscore).
                                 #Replaces them with spaces Example: "hello@world!" → "hello world "

    text=re.sub('https?://\S+|www\.\S+','', text) #Removes URLs https?://\S+ → catches http or https links.
                                           #www\.\S+ → catches www. links  Example: "visit https://example.com" → "visit "
    
    text=re.sub('<.*?>+','', text) #Removes HTML tags. Example: "<p>Hello</p>" → "Hello"
    
    text=re.sub('[%s]' %re.escape(string.punctuation),'', text) #string.puntuation should be string.punctuation (typo).
                                                                #Removes all punctuation characters (.,?!:;'"()[]{} etc.).
    
    text=re.sub('\n','', text) #Removes newline characters. Example: "Hello\nWorld" → "HelloWorld"
   
    text=re.sub('\w*\d\w*','', text)  #Removes any word containing digits.Example: "python3 is cool" → "is cool"
    return text
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
x=data['text']
y=data['class']
```

```python
x_train , x_test, y_train , y_test=train_test_split(x,y, test_size=0.25)  #x → Features (independent variables, e.g., input data).
                                              #y → Labels/targets (dependent variable, e.g., output class).
             #test_size=0.25 → means 25% of the data goes to the test set, and the remaining 75% goes to the training set.
```

## FEATURE EXTRACTION

```python
from sklearn.feature_extraction.text import TfidfVectorizer  

#TfidfVectorizer converts a collection of text documents into a matrix of TF-IDF features.
#Each word in the dataset gets a numerical weight that reflects:
#Term Frequency (TF): How often a word appears in a document.
#Inverse Document Frequency (IDF): How unique that word is across all documents (rare words get higher weight).

vectorization = TfidfVectorizer()  #Creates an instance of TfidfVectorizer with default settings.
xv_train = vectorization.fit_transform(x_train)  

#fit_transform() does two things:
#fit(): Learns the vocabulary and IDF values from the training data (x_train).
#transform(): Converts each document in x_train into a sparse vector of TF-IDF scores.
#If x_train has 5000 documents and vectorizer learns 10,000 unique words, then xv_train shape = (5000, 10000).

xv_test = vectorization.transform(x_test)  #Applies the same vocabulary and IDF values learned from training data to transform the test dataset.
```

## LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))  #This line prints a classification report for your model’s predictions. 
 #It shows metrics like precision, recall, f1-score, and support for each class (fake/real news). Basically, it tells you how well your model did for each label.
```

```python
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not Fake News"
        
def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_xv_test = vectorization.transform(new_def_test["text"])

    # Predictions
    pred_LR = output_label(LR.predict(new_xv_test)[0])
    pred_DT = output_label(DT.predict(new_xv_test)[0])
    pred_GB = output_label(GB.predict(new_xv_test)[0])
    pred_RF = output_label(RF.predict(new_xv_test)[0])

    # Collect results in a dictionary
    results = {
        "Logistic Regression": pred_LR,
        "Decision Tree": pred_DT,
        "Gradient Boosting": pred_GB,
        "Random Forest": pred_RF
    }

    # Majority Voting (final decision)
    from collections import Counter
    final_prediction = Counter(results.values()).most_common(1)[0][0]

    # Print in table format
    print("\n--- Model Predictions ---")
    for model, prediction in results.items():
        print(f"{model:20} : {prediction}")

    print("\nFinal Decision (Majority Vote):", final_prediction)
```
